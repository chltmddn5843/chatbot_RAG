```python
from langchain_community.vectorstores import Chroma
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_core.documents import Document

# 1. ì§ˆë¬¸ìë‹˜ì´ ì‚¬ìš©í•˜ì‹œëŠ” Nomic ëª¨ë¸ ì„¤ì •
# nomic-v2-moe ëª¨ë¸ì€ trust_remote_code=Trueê°€ í•„ìˆ˜ì…ë‹ˆë‹¤.
model_name = "nomic-ai/nomic-embed-text-v2-moe"
encode_kwargs = {'normalize_embeddings': True}

# 2. ì„ë² ë”© ì—”ì§„ ìƒì„±
# query_instructionì„ í†µí•´ ê²€ìƒ‰ ì‹œ 'search_query: ' ì ‘ë‘ì‚¬ê°€ ë¶™ë„ë¡ ì„¤ì •í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
embeddings = HuggingFaceEmbeddings(
    model_name=model_name,
    model_kwargs={'trust_remote_code': True},
    encode_kwargs=encode_kwargs
)

# 3. ì‹¤ìŠµ ë°ì´í„° ì¤€ë¹„
documents = [
    Document(page_content="íŒŒì´ì¬ì€ í”„ë¡œê·¸ë˜ë° ì–¸ì–´ì…ë‹ˆë‹¤."),
    Document(page_content="LangChainì€ LLM í”„ë ˆì„ì›Œí¬ì…ë‹ˆë‹¤."),
    Document(page_content="RAGëŠ” ê²€ìƒ‰ ì¦ê°• ìƒì„±ì…ë‹ˆë‹¤."),
    Document(page_content="ë²¡í„° DBëŠ” ì„ë² ë”©ì„ ì €ì¥í•©ë‹ˆë‹¤."),
]

# 4. ë²¡í„° ì €ì¥ì†Œ(Chroma) ìƒì„± ë° ë°ì´í„° ì €ì¥
# ë³„ë„ì˜ ê²½ë¡œë¥¼ ì§€ì •í•˜ì§€ ì•Šìœ¼ë©´ ë©”ëª¨ë¦¬ ìƒì— ì„ì‹œë¡œ ì €ì¥ë©ë‹ˆë‹¤.
vectorstore = Chroma.from_documents(documents, embeddings)

# 5. Retriever(ê²€ìƒ‰ê¸°)ë¡œ ë³€í™˜
retriever = vectorstore.as_retriever(search_kwargs={"k": 2}) # ìƒìœ„ 2ê°œ ê²°ê³¼ ë°˜í™˜

# 6. ê²€ìƒ‰ ì‹¤í–‰
# ê²€ìƒ‰ ì‹œ ëª¨ë¸ì´ ì§ˆë¬¸ì„ì„ ì•Œ ìˆ˜ ìˆë„ë¡ "search_query: "ë¥¼ ë¶™ì—¬ì¤ë‹ˆë‹¤.
query = "AI ê°œë°œ ë„êµ¬"
results = retriever.invoke(f"search_query: {query}")

print(f"ğŸ” '{query}' ê²€ìƒ‰ ê²°ê³¼:\n")
for i, doc in enumerate(results, 1):
    print(f"[{i}] {doc.page_content}")
```
